```python
###correct code
import math
import numpy as np

def d_t(df_d,feature,l=0):
    
    #pure node-- only class in op
    set_=list(set(df_d['op']))
    if len(set_)==1:
        print('level',l)
        e=entropy_(df_d['op'])
        print('entropy',e)
        print('count of ',set_[0],'is ',len(df_d['op']))
        print('pure node')
        print()
        return
    
    #if no feature left
    elif len(feature)==0:
        print('no features left')
        return
    
    else:
    #decide feature to split on
        max_gain=-1
        #go to all features and pick feature with maximum gain
        for i in feature:
            new_x=df_d[i]
            #calculate gain
            g=gain_ratio(new_x)
            ig=info_gain_(new_x)
            if ig>max_gain:   ###compare their information gains and pick feature with max info gain
                max_gain=ig
                gr=g
                f=i
        e=entropy_(df_d['op'])
        print('level',l)
        print('entropy',e)
        
        #print different values of the feature and their count
        d={}
        for i in df_d['op']:
            d[i]=d.get(i,0)+1
        for i in d:
            print('Count of ',i,'is',d[i])
    
        print('splitting on feature',f,gr)
        
        ###remove this feature from final feature list
        feature.remove(f)
        print()
        
        ### go to all possible values of that particular feature
        t=df_d[f]
        
        for i in set(t):
            x=df_d[df_d[f]==i]
            d_t(x,feature,l+1)

def pi(x):
    d={}
    for i in x:
        d[i]=d.get(i,0)+1
    n=len(x)
    d2={}
    for i in d:
        d2[i]=d[i]/n
    return d2

def info_gain_(x):
    d2=pi(x)
    info_n=0
    for i in d2:
        info_n+=(-1*d2[i]*math.log(d2[i]))
    info_f=0
    for i in d2:
        info_f+=(-1*d2[i]*d2[i]*math.log(d2[i]))
    return info_n-info_f

def entropy_(x):
    d2=pi(x)
    e=0
    for i in d2:
        e+=(-1*d2[i]*math.log(d2[i]))
    return e

def gain_ratio(x):
    info_g=info_gain_(x)
    split_info=split_info_(x)
    return (info_g)/split_info

def split_info_(x):
    split_info=0
    d2=pi(x)
    for i in d2:
        split_info+=(-1*d2[i]*math.log(d2[i]))
    return split_info
```

```python
import pandas as pd
arr=np.array([[True,True],[False,True],[True,False],[False,False]])
arr2=np.array([True,True,True,False])
f=['x1','x2','op']
df=pd.DataFrame(data=np.c_[arr,arr2],columns=f)
d_t(df,f)
```

```python
iris=datasets.load_iris()
features=iris.feature_names
x=iris.data
y=iris.target
features.append('op')
df=pd.DataFrame(np.c_[x,y],columns=features)
df.head()
```

```python
def category(data, *boundaries):
    if data <= boundaries[0]:
        return 'a'
    elif data <= boundaries[1]:
        return 'b'
    elif data <= boundaries[2]:
        return 'c'
    else:
        return 'd'
    
for i in features:
    minimum = df[i].min()    #minimum
    second = df[i].mean()    #mean
    maximum = df[i].max()    #maximum
    
    first = (second + minimum)/2
    third = (second + maximum)/2
    
    df[i] = df[i].apply(category, args = (first, second, third))
df.head()
```
